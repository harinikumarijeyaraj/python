```python
32.List Flattening

Input: [[1, 2], [3, 4], [5]]  
Output: [1, 2, 3, 4, 5]

l = [[1, 2], [3, 4], [5]]
k = []
for i in l:
    for j in i:
        k.append(j)

print(k)
```

```python
33.Find All Numbers That Are Palindromes in List
Input: [121, 133, 123, 88, 97]  
Output: [121, 133, 88]

h = [121, 133, 123, 88, 97]
for b in h:
    if str(b) == str(b)[::-1]:
        print(b)
```

```python
34.Find Longest Word in Sentence

sentence = "Python is fantastic and powerful"
words = sentence.split()
longest = ""

for i in words:
    if len(i) > len(longest):
        longest = i

print("Longest word:", longest)
```

```python
36.Simple Contact Book (Dict based)
Input: {"Ravi": "9876543210", "Anu": "9911223344"}  
Search: "Anu"  
Output: "Contact: 9911223344"

contacts = {"Ravi": "9876543210", "Anu": "9911223344"}
name = "Anu"

if name in contacts:
    print("Contact:", contacts[name])
else:
    print("Name not found")
```

```python

37. Marks to Grade Conversion for N Students

Input: [85, 40, 77, 92, 60]  
Output: ['A', 'F', 'B', 'A', 'C']


marks = [85, 40, 77, 92, 60]
for a in marks:
    if a >= 80:
        print("A")
    elif a >= 70:
        print("B")
    elif a >= 50:
        print("C")
    else:
        print("F")
```
